```python
import pandas as pd
```

```python
df = pd.read_csv("spare_table.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df["Date"] = df["Date"].str.replace("/","-")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df['order_id'] = df.index
```

```python
df.head()
```

```python
date_dim = df[['Date']].drop_duplicates().reset_index(drop=True)
date_dim['date_id'] = date_dim.index
date_dim['purchase_day'] = date_dim['Date'].dt.day
date_dim['purchase_month'] = date_dim['Date'].dt.month
date_dim['purchase_year'] = date_dim['Date'].dt.year

date_dim = date_dim[['date_id','Date','purchase_day','purchase_month','purchase_year']]
date_dim.head()
```

```python
time_dim = df[["Time"]].drop_duplicates().reset_index(drop=True)
time_dim["time_id"] = time_dim.index
time_dim["purchase_hour"] = time_dim["Time"].str.split(":").str[0]
time_dim = time_dim[["time_id", "Time", "purchase_hour"]]

time_dim.head()
```

```python
category_dim = df[['Category']].drop_duplicates().reset_index(drop=True)
category_dim["category_id"] = category_dim.index
category_dim = category_dim[["category_id", "Category"]]
```

```python
category_dim.head()
```

```python
type_dim = df[["Type"]].drop_duplicates().reset_index(drop=True)
type_dim["type_id"] = type_dim.index
type_dim = type_dim[["type_id", "Type"]]
```

```python
type_dim.head()
```

```python
maker_dim = df[["Maker"]].drop_duplicates().reset_index(drop=True)
maker_dim["maker_id"] = maker_dim.index
maker_dim = maker_dim[["maker_id", "Maker"]]
```

```python
maker_dim.head()
```

```python
brand_series_dim = df[["Brand_or_series"]].drop_duplicates().reset_index(drop=True)
brand_series_dim["brand_series_id"] = brand_series_dim.index
brand_series_dim = brand_series_dim[["brand_series_id", "Brand_or_series"]]
```

```python
brand_series_dim.head()
```

```python
spare_dim = df[["Spares"]].drop_duplicates().reset_index(drop=True)
spare_dim["spare_id"] = spare_dim.index
spare_dim = spare_dim[["spare_id", "Spares"]]
```

```python
spare_dim.head()
```

```python
payment_type_dim = df[["Payment"]].drop_duplicates().reset_index(drop=True)
payment_type_dim["payment_type_id"] = payment_type_dim.index
payment_type_dim = payment_type_dim[["payment_type_id", "Payment"]]

payment_type_dim.head()
```

```python
fact_table = df.merge(category_dim, on='Category')\
            .merge(type_dim, on='Type')\
            .merge(maker_dim, on='Maker')\
            .merge(brand_series_dim, on='Brand_or_series')\
            .merge(spare_dim, on='Spares')\
            .merge(date_dim, on='Date')\
            .merge(time_dim, on='Time')\
            .merge(payment_type_dim, on='Payment')\
            [['order_id', 'ID', 'category_id', 'type_id', 'maker_id', 'brand_series_id', 'spare_id', 'date_id', 'time_id', 
              'payment_type_id','Price', 'Quantity', 'Tax(10 percent)', 'Total', 'cogs', 'gross margin percentage', 
              'gross income']]

fact_table
```
