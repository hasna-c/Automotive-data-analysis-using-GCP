```python
import pandas as pd
```

```python
df = pd.read_csv('car_sales.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df = df.dropna(subset=['Reg_year'])
```

```python
df['Reg_year'] = df['Reg_year'].fillna(0).astype(int)
```

```python
df['Reg_year'] = df['Reg_year'].astype(int)
```

```python
df.info()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df['id'] = df.index
```

```python
df.head()
```

```python
maker_dim = df[["Maker"]].drop_duplicates().reset_index(drop=True)
maker_dim["maker_id"] = maker_dim.index
maker_dim = maker_dim[["maker_id", "Maker"]]
```

```python
maker_dim.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
genmodel_dim = df[["Genmodel_ID", "Genmodel"]].drop_duplicates().reset_index(drop=True)
genmodel_dim["Gen_ID"] = genmodel_dim.index
genmodel_dim = genmodel_dim[["Gen_ID", "Genmodel"]]
```

```python
genmodel_dim.head()
```

```python
adv_year_dim = df[["Adv_year"]].drop_duplicates().reset_index(drop=True)
adv_year_dim["Adv_year_ID"] = adv_year_dim.index
adv_year_dim = adv_year_dim[["Adv_year_ID", "Adv_year"]]
```

```python
adv_year_dim.head()
```

```python
month_mapping = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}

# Convert 'Adv_month' column to month names
df['Adv_month'] = df['Adv_month'].map(month_mapping)
```

```python
adv_month_dim = df[["Adv_month"]].drop_duplicates().reset_index(drop=True)
adv_month_dim["Adv_month_ID"] = adv_month_dim.index
adv_month_dim = adv_month_dim[["Adv_month_ID", "Adv_month"]]
```

```python
adv_month_dim.head()
```

```python
color_dim = df[["Color"]].drop_duplicates().reset_index(drop=True)
color_dim["Color_ID"] = color_dim.index
color_dim = color_dim[["Color_ID", "Color"]]
```

```python
color_dim.head()
```

```python
reg_year_dim = df[["Reg_year"]].drop_duplicates().reset_index(drop=True)
reg_year_dim["Reg_year_ID"] = reg_year_dim.index
reg_year_dim = reg_year_dim[["Reg_year_ID", "Reg_year"]]
```

```python
reg_year_dim.head()
```

```python
fuel_type_dim = df[["Fuel_type"]].drop_duplicates().reset_index(drop=True)
fuel_type_dim["Fuel_type_ID"] = fuel_type_dim.index
fuel_type_dim = fuel_type_dim[["Fuel_type_ID", "Fuel_type"]]
```

```python
fuel_type_dim.head()
```

```python
bodytype_dim = df[["Bodytype"]].drop_duplicates().reset_index(drop=True)
bodytype_dim["Bodytype_ID"] = bodytype_dim.index
bodytype_dim = bodytype_dim[["Bodytype_ID", "Bodytype"]]
```

```python
bodytype_dim.head()
```

```python
gearbox_dim = df[["Gearbox"]].drop_duplicates().reset_index(drop=True)
gearbox_dim["Gearbox_ID"] = gearbox_dim.index
gearbox_dim = gearbox_dim[["Gearbox_ID", "Gearbox"]]
```

```python
gearbox_dim.head()
```

```python
fact_table = df.merge(maker_dim, on='Maker')\
            .merge(genmodel_dim, on='Genmodel')\
            .merge(adv_year_dim, on='Adv_year')\
            .merge(adv_month_dim, on='Adv_month')\
            .merge(color_dim, on='Color')\
            .merge(reg_year_dim, on='Reg_year')\
            .merge(fuel_type_dim, on='Fuel_type')\
            .merge(bodytype_dim, on='Bodytype')\
            .merge(gearbox_dim, on='Gearbox')\
            [['id', 'Adv_ID', 'maker_id', 'Gen_ID', 'Adv_year_ID', 'Adv_month_ID', 'Color_ID', 'Reg_year_ID', 'Fuel_type_ID', 
              'Bodytype_ID','Gearbox_ID', 'Runned_Miles', 'Engin_size', 'Price']]

fact_table
```
